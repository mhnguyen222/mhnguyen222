```python
def get_hs_regents_subject_by_year_for_borough(subject, year, borough):
  return df[(df['School Level'] == 'High school') & (df['Regents Exam'] == subject) & (df['Year'] == year) & (df['Borough'] == borough)].copy()
```

```python
manhattan_scores = get_hs_regents_subject_by_year_for_borough('Common Core Algebra', 2017, 'M')
```

```python
queens_scores = get_hs_regents_subject_by_year_for_borough('Common Core Algebra', 2017, 'Q')
```

```python
manhattan_scores['Percent Scoring 80 or Above'] = manhattan_scores['Percent Scoring 80 or Above'].map(float)

```

```python
queens_scores['Percent Scoring 80 or Above'] = queens_scores['Percent Scoring 80 or Above'].map(float)
```

```python
def convert_80_above_to_float(my_df):
  my_df['Percent Scoring 80 or Above'] = my_df['Percent Scoring 80 or Above'].map(float)
```

```python
def process_borough(borough):
   result = get_hs_regents_subject_by_year_for_borough('Common Core Algebra', 2017, borough)
   convert_80_above_to_float(result)
   return result['Percent Scoring 80 or Above'].mean()

```

```python
borough_mapping = { 'M': 'Manhattan', 'X': 'Bronx', 'K': 'Brooklyn', 'Q' : 'Queens', 'R': 'Staten Island'}
```

```python
data = dict((map(lambda x: (borough_mapping[x], process_borough(x)), df['Borough'].drop_duplicates().values)))
```

```python
graph = plt.bar(data.keys(), data.values(), color = 'maroon', width = 0.4)
```
