```python
##Question: What is the leading cause of death and its correlations?
```

```python
##Clean data with just focusing on the latest dates, cause, sex, race, deaths, and deathrates.
```

```python
new_df = df [['year','leading_cause','sex','race_ethnicity','deaths','death_rate']].copy()
```

```python
new_df.head()
```

```python
# Check for missing values in the dataset
print("Missing values in the dataset:")
print(new_df.isnull().sum())

# Drop rows with missing values
new_df = new_df.dropna()
```

```python
# Convert 'year' to datetime type
new_df['year'] = pd.to_datetime(new_df['year'], format='%Y')
```

```python
# Filter data for the latest year
latest_year = new_df['year'].max()
latest_data = new_df[new_df['year'] == latest_year]

# Explore the unique values in 'leading_cause'
print("Unique leading causes of death:")
print(latest_data['leading_cause'].unique())
```

```python
# Exploratory Data Analysis (EDA)
# Analyze the distribution of leading causes of death
plt.figure(figsize=(14, 8))
sns.countplot(y='leading_cause', data=latest_data, order=latest_data['leading_cause'].value_counts().index, palette='muted')
plt.title(f"Distribution of Leading Causes of Death in {latest_year.year}")
plt.xlabel('Number of Occurrences')
plt.ylabel('Leading Cause of Death')
plt.show()

# Statistical analysis: Group by leading cause to find the most common cause
common_cause = latest_data.groupby('leading_cause')['deaths'].sum().sort_values(ascending=False).head(1)
print(f"The most common leading cause of death in {latest_year.year} is: {common_cause.index[0]}")
```

```python
# Visualization: Pie chart for the distribution of deaths by sex
plt.figure(figsize=(10, 8))
latest_data['sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title(f"Distribution of Deaths by Sex in {latest_year.year}")
plt.ylabel('')
plt.show()
```

```python
# Correlation analysis: Scatter plot for deaths vs death rate
plt.figure(figsize=(12, 6))
sns.scatterplot(x='deaths', y='death_rate', data=latest_data, hue='sex', palette='Set2')
plt.title(f"Scatter Plot of Deaths vs Death Rate in {latest_year.year}")
plt.xlabel('Number of Deaths')
plt.ylabel('Death Rate')
plt.show()
```

```python
# Convert 'deaths' column to numeric (in case it's not already)
latest_data['deaths'] = pd.to_numeric(latest_data['deaths'], errors='coerce')

# Visualization: Bar chart for leading causes of death by race and gender
plt.figure(figsize=(16, 10))
sns.barplot(x='deaths', y='leading_cause', hue='sex', data=latest_data, palette='muted', ci=None)
plt.title(f"Leading Causes of Death in {latest_year.year} by Race and Gender")
plt.xlabel('Number of Deaths')
plt.ylabel('Leading Cause of Death')
plt.legend(title='Gender', loc='upper right')
plt.show()
```

```python
# Convert 'leading_cause' column to string
df['leading_cause'] = df['leading_cause'].astype(str)
```

```python

# Group by race and find the highest leading cause for each race
highest_cause_by_race = df.groupby('race_ethnicity')['leading_cause'].apply(lambda x: x.value_counts().idxmax()).reset_index(name='highest_cause')

# Plotting
plt.figure(figsize=(12, 8))
sns.countplot(x='race_ethnicity', hue='highest_cause', data=highest_cause_by_race, palette='viridis')
plt.title("Highest Leading Cause of Death for Each Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.legend(title='Leading Cause', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
